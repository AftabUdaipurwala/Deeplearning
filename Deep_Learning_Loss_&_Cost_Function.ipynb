```python
y_predicted = np.array([1,1,0,0,1])
y_actual = np.array([0.3,0.7,1,0,0.5])
```

```python
# Creating the function using the for loop 
def mae(y_predicted, y_actual):
  total_error=0
  for yt, yp in zip(y_actual,y_predicted):
    total_error = abs(yt-yp)
  print('total_error :',total_error)
  print('mae :', total_error/len(y_actual))
```

```python
mae(y_predicted, y_actual)
```

```python
# Doing the same thing using Numpy Array
print("MAE :",np.mean(np.abs(y_predicted-y_actual)))
```

```python
# Doing the MSE thing using Numpy Array
difference_array = np.subtract(y_predicted, y_actual)
squared_array = np.square(difference_array)
mse = squared_array.mean()
print('MSE :',mse)
```

## ***Log loss or Binary Cross entropy function***

```python
epsilon = 1e-15
y_predicted_new = [max(i,epsilon) for i in y_predicted ]
y_predicted_new
```

```python
# Log loss or Binary Cross entropy function
epsilon = 1e-15
y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new ]
y_predicted_new
```

```python
np.log(y_predicted_new)
```

```python
# Calculating the log loss using this function

-np.mean(y_actual*np.log(y_predicted_new)+(1-y_actual)*np.log(1-y_predicted_new))
```
