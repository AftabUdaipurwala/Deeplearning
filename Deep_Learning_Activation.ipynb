```python
# Sigmoid function
import math
def sigmoid(x):
  return 1/(1+math.exp(-x))
```

```python
sigmoid(1)
```

```python
def tanh(x):
  return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh(-5)
```

```python
tanh(5)
```

```python
def relu(x):
  return max(0,x)
```

```python
relu(6)
```

```python
relu(-9)
```

```python
def leakyrelu(x):
  return max(0.1*x,x)
```

```python
leakyrelu(-10)
```

```python
leakyrelu(10)
```
